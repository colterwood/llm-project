```python
import pandas as pd
ds_train = pd.DataFrame(ds['train'])
ds_test = pd.DataFrame(ds['test'])
```

```python
ds_train.head()
```

```python
ds_test.head()
```

```python
from datasets import Dataset, DatasetDict
# assign the splits
train = Dataset.from_pandas(ds_train)
test = Dataset.from_pandas(ds_test)
# reconstruct both datasets into a Dataset Dict object
new_ds = DatasetDict(
    {
        'train': train,
        'test': test
    }
)
# view the resulting dataset dict object
new_ds
```

```python
from datasets import ClassLabel

# Define ClassLabel feature
label_feature = ClassLabel(names=['neg', 'pos'])

# Function to apply label assignment
def assign_label(example):
    example['label'] = 'pos' if example['score'] > 6.5 else 'neg'
    return example

# Apply the function to each split and cast the column
new_ds = new_ds.map(assign_label)
new_ds = new_ds.cast_column('label', label_feature)

# Print the updated dataset structure
print(new_ds)
```

```python
import numpy as np

median_threshold = np.median(new_ds["train"]["score"])
print(f"Median threshold: {median_threshold}")
```

```python
from datasets import ClassLabel

# Define ClassLabel feature
label_feature = ClassLabel(names=['neg', 'pos'])

# Function to apply label assignment
def assign_label(example):
    example['label'] = 'pos' if example['score'] >= median_threshold else 'neg'
    return example

# Apply the function to each split and cast the column
new_ds = new_ds.map(assign_label)
new_ds = new_ds.cast_column('label', label_feature)

# Print the updated dataset structure
print(new_ds)
```
